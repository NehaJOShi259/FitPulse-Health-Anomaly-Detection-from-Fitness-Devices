```python
import pandas as pd
import numpy as np
```

```python
activity = pd.read_csv("data/dailyActivity_merged.csv")
heart = pd.read_csv("data/heartrate_seconds_sample.csv")
sleep = pd.read_csv("data/sleepDay_merged.csv")
```

```python
# Activity
activity['ActivityDate'] = pd.to_datetime(activity['ActivityDate'])
activity.set_index('ActivityDate', inplace=True)

# Heart rate
heart['Time'] = pd.to_datetime(heart['Time'])
heart.set_index('Time', inplace=True)

# Sleep
sleep['SleepDay'] = pd.to_datetime(sleep['SleepDay'])
sleep.set_index('SleepDay', inplace=True)
```

```python
# Resample using forward-fill for missing values
activity_1min = activity.resample('1T').ffill()
heart_1min = heart.resample('1T').mean().ffill()
sleep_1min = sleep.resample('1T').ffill()
```

```python
# Keep only the first occurrence for duplicate timestamps
activity = activity[~activity.index.duplicated(keep='first')]
heart = heart[~heart.index.duplicated(keep='first')]
sleep = sleep[~sleep.index.duplicated(keep='first')]
```

```python
# Resample using 1-minute intervals
activity_1min = activity.resample('1min').ffill()
heart_1min = heart.resample('1min').mean().ffill()
sleep_1min = sleep.resample('1min').ffill()
```

```python
merged = activity_1min.join(heart_1min, how='outer').join(sleep_1min, how='outer')
merged.fillna(method='ffill', inplace=True)

merged.head()
```
